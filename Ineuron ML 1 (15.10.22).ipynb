```python
# Independent and Dependent features
```

```python
X = dataset.iloc[ : , : -1]
```

```python
y = dataset.iloc[ : , -1]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=10)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
# Standardize or feature scaling the dataset
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test) # use transform to avoid data leakage
```

```python
X_train
```

```python
X_test
```

```python
## Model training
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression() # create object
```

```python
reg
```

```python
reg.fit(X_train, y_train) # train model
```

```python
# print the co-efficient(13) and intercept(1)
```

```python
print(reg.coef_) # if my all features i.e. 13 are 0 than my predicted value is 0 i.e. house price
```

```python
print(reg.intercept_)# my predicted value is 0 i.e. house price
```

```python
# prediction of the test data
```

```python
reg_pred = reg.predict(X_test)
```

```python
reg_pred
```

```python
## Assumptions of linear Regression
```

```python
plt.scatter(y_test, reg_pred) # y_test is truth point & reg_pred is assumption points
plt.xlabel("Test Truth Data")
plt.ylabel("Test predicted Data")
```

```python
# residual (y_test-reg_pred) difference error between test and predicted data point error
residual = y_test - reg_pred
```

```python
residual
```

```python
sns.displot(residual, kind = 'kde') # kernel density astimator i.e. kde
```

```python
# Scatter plot with prediction and residual called Uniform distribution
```

```python
plt.scatter(reg_pred, residual) # if get uniform distribution than model is good
```

```python
# Performance metrics
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test, reg_pred))
print(mean_absolute_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
# R square and adjusted R square
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test, reg_pred)
print(score)
```

```python
# Adjusted R square error
```

```python
1 - (1 - score)* (len(y_test)-1) / (len(y_test) - X_test.shape[1] -1)
```
